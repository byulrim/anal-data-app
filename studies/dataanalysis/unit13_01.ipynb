
<span style="color: #E0E0E0; font-size: 15px">
일반적으로 히스토그램은 연속적인 데이터를 표현하기 때문에 막대 사이의 간격이 없는 것이 정석입니다.  

하지만 **주사위 눈**이나 **설문조사 점수**처럼 **딱 떨어지는 숫자를 시각화할 때**는 막대가 붙어 있으면 답답해 보일 수 있습니다.

- rwidth=1.0 (기본): 막대들이 다닥다닥 붙어 있어 연속적인 느낌을 줍니다.
- rwidth=0.8: 막대 사이에 약간의 틈이 생겨 각 항목(주사위 눈 1, 2, 3...)이 개별적으로 명확하게 구분됩니다.

- 중앙 정렬: bins를 [0.5, 1.5, ...]와 같이 설정하면, 숫자 1이 0.5~1.5 구간의 딱 한가운데 놓이게 되어 보기에 매우 편안합니다.
- 빈 틈 방지: 단순히 bins=6으로 설정하면, 프로그램이 최소값과 최대값을 기준으로 구간을 나누면서 특정 숫자가 옆 구간에 걸치거나 막대 두께가 제각각이 되는 현상이 발생할 수 있습니다.
- rwidth 활용: rwidth=0.8 같은 옵션을 추가하면 막대 사이에 살짝 간격이 생겨서, 각 주사위 눈의 빈도를 개별적으로 확인하기 훨씬 좋아집니다.

- `rwidth`는 Relative Width(상대적 너비)의 약자로, 히스토그램 막대의 너비를 전체 구간 너비에 대한 비율로 조절하는 파라미터입니다.
- 쉽게 말해, "막대를 얼마나 뚱뚱하게 혹은 날씬하게 만들지" 결정하며, 막대 사이의 간격(Gap)을 조절하는 용도로 주로 사용됩니다.
</span>

```python
dice = np.random.choice(range(1, 7), 1000000, p=[0.1, 0.2, 0.3, 0.2, 0.1, 0.1])
# print(dice)

# plt.hist(dice, bins=[0.5, 1.5, 2.5, 3.5, 4.5, 5.5, 6.5], rwidth=0.8)
edges = np.arange(0.5, 7.5, 1.)
# print(edges)

plt.hist(dice, bins=edges, rwidth=0.7)
plt.show()
```

```python
dice = [random.choice(range(1, 7)) for _ in range(1000000)]
# print(dice)

# edges = np.arange(0.5, 7.5, 1.)
edges = [x + 0.5 for x in range(7)]

plt.hist(dice, bins=edges, rwidth=0.8)
plt.show()
```

<span style="font-size: 15px">
배열(벡터/행렬)과 스칼라의 연산은 스칼라 값을 배열의 모든 요소에 동일하게 적용하는 것으로, 

 - 스칼라배(scalar multiplication)라고 합니다.  
 - 사칙연산(+, -, *, /) 모두 배열의 각 원소가 스칼라와 연산되어 원래 배열과 같은 크기의 결과 배열을 반환합니다.  

 - 사용처: 데이터 정규화, 스케일링, 행렬 전체의 값 증가/감소 등 데이터 분석 및 딥러닝에서 필수적으로 사용됩니다.
</span>

```python
a = np.arange(-np.pi, np.pi, np.pi / 100)

plt.plot(a, np.sin(a), color="red", label="sin")
plt.plot(a, np.cos(a), color="blue", label="cos")
plt.plot(np.pi/2 + a, np.sin(a), color="green", label="sin shifted")

plt.show()
```

```python
a = np.arange(-np.pi, np.pi, np.pi / 100)
a_edges = np.linspace(-np.pi, np.pi, 20)

fig, ax = plt.subplots(2, 2, figsize=(10, 8))

ax[0, 0].hist(a, bins=a_edges, label="histogram")
ax[0, 0].set_title("Histogram", fontsize=10)

ax[0, 1].plot(a, np.sin(a), label="sin")
ax[0, 1].set_title("Sine", fontsize=10)

ax[1, 0].plot(a, np.cos(a), label="cos")
ax[1, 0].set_title("Cosine", fontsize=10)

ax[1, 1].plot(np.pi / 2 + a, np.sin(a), label="Sine shifted")
ax[1, 1].set_title("Sine shifited", fontsize=10)

plt.show()
```

```python
a = np.arange(-5, 5)
print(a, type(a), a.dtype, a.shape)
print(a<0)
print(a[a<0])

mask1 = abs(a) > 3 # boolean masking - 불리언 값으로 배열과 같은 크기의 배열 생성
print(a[mask1]) # boolean indexing - 불리언 값을 이용하여 배열에서 원하는 값을 인덱싱하여 추출
```

```python
a = np.arange(-5, 5)
mask1 = abs(a) > 3
mask2 = abs(a) % 2 == 0

print(a[mask1 | mask2])
print(a[mask1 & mask2])
```

```python
x = np.random.randint(-100, 100, 1000)
y = np.random.randint(-100, 100, 1000)
size = np.random.rand(1000) * 100

print(x.shape, y.shape, size.shape)

mask1 = abs(x) > 50
mask2 = abs(y) > 50

x = x[mask1 | mask2]
y = y[mask1 | mask2]
size = size[mask1 | mask2]

print(x.shape, y.shape, size.shape)

plt.scatter(x, y, s=size, c=x, cmap="jet", alpha=0.7)
plt.colorbar()
plt.show()
```
