```python
data_files = glob("./data/*.csv")

for file in data_files:
    print(f"data_files[{data_files.index(file)}]: {file}")
```

```python
with open(data_files[1], "r", encoding="utf-8") as data_file:
    data = csv.reader(data_file)
    header = next(data)

    start = 0
    total_col_num = len(header)
    for col in header:
        print(f"{start: >3} {(total_col_num - start)* -1: >3}: {col}")
        start += 1
```

```python
dong_name = "세류2동"
header = ""

male = []
female = []

with open(data_files[1], "r", encoding="utf-8") as data_file:
    data = csv.reader(data_file)

    header = next(data)

    for row in data:
        if dong_name in row[0]:
            for i in range(3, 104, 1):
                male.append(int(row[i]))
                female.append(int(row[i + 103]))
            break
# print(male, female, sep="\n")
```

```python
plt.figure(figsize=(10, 5))
plt.title("Population of Seoryu2-dong")
plt.plot(range(101), male, color="skyblue", label="male", marker=".")
plt.plot(range(101), female, color="hotpink", label="female", marker=".")
plt.legend()
plt.show()
```

```python
header = []
result = []
dong_name = "세류2동"

with open(data_files[1], "r", encoding="utf-8") as data_file:
    data = csv.reader(data_file)
    
    header = next(data)

    for row in data:
        if dong_name in row[0]:
            for i in range(101):
                result.append(int(row[i + 3]) - int(row[i + 106]))

print(result)
```

```python
plt.figure(figsize=(10, 5))
plt.title(f"{dong_name} 인구 남녀 성비", fontsize=10)
plt.bar(range(len(result)), result, label="남성 인구수 - 여성 인구수")
plt.legend()
plt.show()
```
