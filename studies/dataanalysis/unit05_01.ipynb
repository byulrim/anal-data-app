```python
from glob import glob
import csv
from matplotlib import pyplot as plt

plt.rc("font", family="Malgun Gothic")
plt.rcParams["axes.unicode_minus"] = False
```

```python
file_paths = glob("./data/*.csv")

for path in file_paths:
    print(f"file_paths[{file_paths.index(path)}] : {path}")
```

```python
result = []

with open(file_paths[1], "r", encoding="utf-8") as tempo_file:
    data = csv.reader(tempo_file)

    header = next(data)
    print(header)

    for row in data:
        if row[-1] != "":
            result.append(float(row[-1]))
```

```python
print(result, len(result), sep="\n")
```

```python
plt.figure(figsize=(10, 3))
plt.plot(result, color="red")
plt.show()
```

```python
result = []

with open(file_paths[1], "r", encoding="utf-8") as tempo_file:
    data = csv.reader(tempo_file)
    
    header = next(data)

    for row in data:
        if (row[-1] != "") and (row[0].split("-")[1] == "08"):
            result.append(float(row[-1]))

plt.plot(result, color="hotpink")
plt.show()
```

```python
high_tempo = []
low_tempo = []

with open(file_paths[1], "r", encoding="utf-8") as tempo_file:
    data = csv.reader(tempo_file)
    
    header = next(data)

    for row in data:
        if (row[-1] != "" and row[-2] != "") \
            and (int(row[0].split("-")[0]) >= 1973 \
                 and row[0].split("-")[1] == "02" and row[0].split("-")[2] == "09"):
                high_tempo.append(float(row[-1]))
                low_tempo.append(float(row[-2]))

# for i in high_tempo:
#      print(f"최고온도: {i: >5.1f}")
# for i in low_tempo:
#      print(f"최저온도: {i: >5.1f}")


plt.figure(figsize=(7, 5))
plt.title("내 생일의 기온 변화 그래프" , fontsize=10)
plt.plot(high_tempo, color="hotpink", marker=".", label="High Temp")
plt.plot(low_tempo, color="skyblue", marker=".", label="Low Temp")
plt.legend()
plt.show()
```
