```python
data_files = glob("./data/*.csv")

for file in data_files:
    print(f"data_files[{data_files.index(file)}]: {file}")
```

```python
with open(data_files[4], "r", encoding="utf-8-sig") as read_file, \
    open("./data/output.csv", "w", encoding="utf-8", newline="") as write_file:
    reader = csv.reader(read_file)
    writer = csv.writer(write_file)

    header = next(reader)
    del header[-1]
    writer.writerow(header)
    
    for row in reader:
        del row[-1]
        for i in range(0, len(row)):
            if row[i].find(",") != -1:
                row[i] = int(row[i].replace(",", ""))
        writer.writerow(row)
```

```python
header = []

with open(data_files[3], "r", encoding="utf-8") as data_file:
    data = csv.reader(data_file)

    header = next(data)

    for i in header:
        print(f"header[{header.index(i)}]: {i}")
```

```python
max = 0
rate = 0

with open(data_files[3], "r", encoding="utf-8") as data_file:
    data = csv.reader(data_file)

    next(data)

    for row in data:
        for i in range(4, 8):
            row[i] = int(row[i])
        if row[6] != 0:
            rate = row[4] / row[6] # 무임승차 인원으로 유임승차 인원을 나눈 값
            if max < rate:
                max = rate
                print(row, round(rate, 2))
```

```python
max = 0
rate = 0
max_station = ""
with open(data_files[3], "r", encoding="utf-8") as data_file:
    data = csv.reader(data_file)

    next(data)

    for row in data:
        for i in range(4, 8):
            row[i] = int(row[i])

        if (row[4] + row[6]) > 100000:
            rate = row[4] / (row[4] + row[6])

        if max < rate:
            max = rate
            max_station = row[3] + "(" + row[1] + ")"

print(f"유임 승차 비율이 가장 높은 역:\n역이름(노선명): {max_station}, 유임 승차 비율: {max * 100:.2f}")
```

```python
max = [0] * 4
max_station = [""] * 4
label = ["유임승차", "유임하차", "무임승차", "무임하차"]

with open(data_files[3], "r", encoding="utf-8") as data_file:
    data = csv.reader(data_file)

    next(data)

    for row in data:
        for i in range(4, 8):
            row[i] = int(row[i])

            if max[i - 4] < row[i]:
                max[i - 4] = row[i]
                max_station[i - 4] = row[3] + " " + row[1]

for i in range(0, len(max)):
    print(f"{label[i]: <4}: {max_station[i]: <15} ({max[i]: >9,})")
```
