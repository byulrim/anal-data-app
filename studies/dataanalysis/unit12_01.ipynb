```python
import csv
from glob import glob
```

```python
data_files = glob("./data/*.csv")

for data_file in data_files:
    print(f"data_files[{data_files.index(data_file)}]: {data_file}")
```

```python
with open(data_files[6], "r", encoding="utf-8-sig") as read_file, \
    open("./data/subwaytime.csv", "w", newline="", encoding="utf-8") as write_file:
    reader = csv.reader(read_file)
    writer = csv.writer(write_file)

    while reader.line_num < 2:
        header = next(reader)
        del header[-1]
        writer.writerow(header)

    for row in reader:
        del row[-1]
        for i in range(4, len(row)):
            if row[i] != -1:
                row[i] = int(row[i].replace(",", ""))
        writer.writerow(row)
```

```python
header = []

with open(data_files[5], "r", encoding="utf-8-sig") as data_file:
    data = csv.reader(data_file)

    while data.line_num < 2:
        header.append(next(data))

    for row in data:
        row[4:] = map(int, row[4:])
        print(row)
```
