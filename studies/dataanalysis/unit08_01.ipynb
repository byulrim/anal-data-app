```python
import csv
from matplotlib import pyplot as plt
from glob import glob

plt.style.use("seaborn-v0_8-whitegrid")
plt.rcParams["font.family"] = "Malgun Gothic"
plt.rcParams["axes.unicode_minus"] = False
plt.rcParams["font.size"] = 10
```

```python
print(plt.style.available)
```

```python
data_paths = glob("./data/*.csv")

for path in data_paths:
    print(f"data_paths[{data_paths.index(path)}]: {path}")
```

```python
dong_name = input("     ( )  : ")
fq_dong_name = ""
result = []

with open(data_paths[0], "r", encoding="utf-8") as pop_file:
    data = csv.reader(pop_file)

    headers = next(data)

    for row in data:
        if dong_name in row[0]:
            # fq_dong_name = row[0]
            fq_dong_name = row[0][:row[0].index("(")]    
            for col in row[3:]:
                result.append(int(col.replace(",", "")))

    if not result:
        print(f"{dong_name}   .")
    else:
        plt.figure(figsize=(10, 5), dpi=300)
        plt.title(f"\"{fq_dong_name}\"   ", fontsize=10)
        plt.plot(result, marker=".")
        plt.show()

```
