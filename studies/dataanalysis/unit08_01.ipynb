```python
import csv
from matplotlib import pyplot as plt
from glob import glob

plt.style.use("seaborn-v0_8-whitegrid")
plt.rcParams["font.family"] = "Malgun Gothic"
plt.rcParams["axes.unicode_minus"] = False
plt.rcParams["font.size"] = 10
```

```python
print(plt.style.available)
```

```python
data_paths = glob("./data/*.csv")

for path in data_paths:
    print(f"data_paths[{data_paths.index(path)}]: {path}")
```

```python
dong_name = input("인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해 주세요: ")
fq_dong_name = ""
result = []

with open(data_paths[0], "r", encoding="utf-8") as pop_file:
    data = csv.reader(pop_file)

    headers = next(data)

    for row in data:
        if dong_name in row[0]:
            # fq_dong_name = row[0]
            fq_dong_name = fq_dong_name[0: fq_dong_name.index("(")]    
            for col in row[3:]:
                result.append(int(col.replace(",", "")))
    
    if not result:
        print(f"{dong_name} 지역은 없는 지역입니다.")
        exit()


plt.figure(figsize=(10, 5), dpi=300)
plt.title(f"\"{fq_dong_name}\" 지역의 인구 구조", fontsize=10)
plt.plot(result, marker=".")
plt.show()
```
