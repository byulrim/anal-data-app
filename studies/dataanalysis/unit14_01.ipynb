```python
data_files = glob(".//data//*.csv")

for i, data_file in enumerate(data_files):
    print(f"data_files[{i}]: {data_file}")
```

```python
with open(data_files[0], "r", encoding="utf-8-sig") as data_file:
    data = list(csv.reader(data_file))
    data = data[1:]

search_location = "세류2동"
# search_location = input("인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력하세요: ")
home_list = [row[2:] for row in data if search_location in row[0]][0]
home_array = np.array(home_list[1:]).astype(int).flatten() / int(home_list[0].replace(",",""))
print(home_array, home_array.shape)
# away_list = [row[2:] for row in data]
# away_array = np.array(away_list[3:]).astype(int) / int(away_list[2])
# print(home_array)
# print(away_array)
```

```python
plt.figure(figsize=(10, 5))
plt.title(f"{search_location} 지역의 인구 구조")
plt.plot(home_array)
plt.show()
```
