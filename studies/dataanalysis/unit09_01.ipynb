```python
import csv
from matplotlib import pyplot as plt
from glob import glob

plt.style.use("seaborn-v0_8-whitegrid")
plt.rcParams["font.family"] = "Malgun Gothic"
plt.rcParams["axes.unicode_minus"] = False
plt.rcParams["font.size"] = 10
```

```python
data_paths = glob("./data/*.csv")

for path in data_paths:
    print(f"data_paths[{data_paths.index(path)}]: {path}")
```

**고려 사항**
- [해결]**제주특별자치도** 소계만 가져와야 함.
    - 현재는 제주특별자치도 문자열이 있는 모든 행에서 데이터를 누계하여 가져옴.
- **같은 이름의 행정구역** 이 두 개 이상인 경우도 고려 要.

```python
# new code
header = []

female = []
male = []

region_name = "제주특별자치도"

with open(data_paths[1], "tr", encoding="utf-8") as pop_file:
    pop_data = csv.reader(pop_file)

    next(pop_data)

    for row in pop_data:
        if region_name in row[0]:
            for i in range(101):
                male.append(int(row[i + 3].replace(",", "")) * -1)
                female.append(int(row[(i + 1) * -1].replace(",", "")))
            break

female.reverse()

print(female, male, sep="\n")
```

```python
plt.figure(figsize=(10, 5), dpi=300)
plt.title(f"{region_name} 지역의 남녀 성별 인구 분포", fontsize=10)
plt.barh(range(101), male, label="남성")
plt.barh(range(101), female, label="여성")
plt.legend(fontsize=10)
plt.show()
```
