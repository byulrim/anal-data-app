```python
results = []

with open(data_files[1], "r", encoding="utf-8") as temp_file:
    data = csv.reader(temp_file)

    next(data)

    for row in data:
        if row[-1] != "":
            results.append(float(row[-1]))

plt.hist(results, bins=70, color="red")
plt.show()
```

```python
aug = []

with open(data_files[1], "r", encoding="utf-8") as temp_file:
    data = csv.reader(temp_file)

    next(data)

    for row in data:
        if (row[-1] != "") and (row[0].split("-")[1] == "08"):
            aug.append(float(row[-1]))

plt.hist(aug, bins=70, color="red")
plt.show()
```

```python
jan = []
aug = []

with open(data_files[1], "r", encoding="utf-8") as temp_file:
    data = csv.reader(temp_file)

    next(data)

    for row in data:
        if row[-1] != "":
            if row[0].split("-")[1] == "08":
                aug.append(float(row[-1]))
            if row[0].split("-")[1] == "01":
                jan.append(float(row[-1]))

plt.hist(aug, bins=70, color="red", label="Aug")
plt.hist(jan, bins=70, color="blue", label="Jan")
plt.legend()
plt.show()
```

```python
result = []

with open(data_files[1], "r", encoding="utf=8") as temp_file:
    data = csv.reader(temp_file)

    header = next(data)

    for row in data:
        if row[4] != "":
            result.append(float(row[4]))

plt.boxplot(result)
plt.show()
```

```python
jan = []
aug = []

with open(data_files[1], "r", encoding="utf-8") as temp_file:
    data = csv.reader(temp_file)

    header = next(data)

    for row in data:
        month = row[0].split("-")[1]
        if row[4] != "":
            if month == "01":
                jan.append(float(row[4]))
            if month == "08":
                aug.append(float(row[4]))

# plt.boxplot(aug)
# plt.boxplot(jan)
plt.figure(figsize=(5, 5))
plt.boxplot([aug, jan])
plt.show()
```

```python
months = []
for i in range(12):
    months.append([])

with open(data_files[1], "r", encoding="utf-8") as temp_file:
    data = csv.reader(temp_file)

    header = next(data)

    for row in data:
       if row[4] != "":
           months[int(row[0].split("-")[1]) - 1].append(round(float(row[4]) - float(row[3]), 1))

print(months)

plt.boxplot(months)
plt.show()
```

```python
days = []
for i in range(31):
    days.append([])


with open(data_files[1], "r", encoding="utf-8") as temp_file:
    data = csv.reader(temp_file)

    header = next(data)

    for row in data:
       if (row[4] != "") and (row[0].split("-")[1] == "08"):
           days[int(row[0].split("-")[2]) - 1].append(float(row[4]))


plt.figure(figsize=(10, 5), dpi=300)
plt.style.use("ggplot")
plt.boxplot(days, showfliers=False)
plt.show()
```
