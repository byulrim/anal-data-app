```python
from matplotlib import pyplot as plt
import csv
from glob import glob
```

```python
data_files = glob("./data/*.csv")

for file in data_files:
    print(f"data_files[{data_files.index(file)}] : {file}")
```

```python
plt.hist([1, 1, 2, 3, 4, 5, 6, 6, 7, 8, 10])
plt.show()
```

```python
import random

dice = []

for i in range(1000000):
    dice.append(random.randrange(1, 7))

print(len(dice))
plt.hist(dice, bins=6, range=(1, 6))
plt.show()
```

```python
results = []

with open(data_files[1], "r", encoding="utf-8") as temp_file:
    data = csv.reader(temp_file)

    next(data)

    for row in data:
        if row[-1] != "":
            results.append(float(row[-1]))

plt.hist(results, bins=70, color="red")
plt.show()
```

```python
aug = []

with open(data_files[1], "r", encoding="utf-8") as temp_file:
    data = csv.reader(temp_file)

    next(data)

    for row in data:
        if (row[-1] != "") and (row[0].split("-")[1] == "08"):
            aug.append(float(row[-1]))

plt.hist(aug, bins=70, color="red")
plt.show()
```
