```python
scores = [38, 37, 7, 16, 95, 71, 63, 48, 49, 66, 37]
grades = [""] * len(scores)

for i, score in enumerate(scores):
    if score >= 90:
        grade = "A"
    elif score >= 80:
        grade = "B"
    elif score >=70:
        grade = "C"
    elif score >= 60:
        grade = "D"
    else:
        grade ="F"
    
    grades[i] = grade

for score, grade in zip(scores, grades):
    print(f"{score: >2}: {grade}")
    
```

```python
def print_multiplication_table():
    print("구구단 몇 단을 계산할까요?")

    # table = int(input())
    try:
        table = int(input())
    except ValueError:
        print("정수만 입력해 주세요.")
        return -1

    print(f"구구단 {table}단을 계산합니다.")

    # for i in range(1, 10):
    #     print(f"{table: ,} X {i} = {table * i:2d}")

    results = [f"{table: ,} X {i} = {table * i:2d}" for i in range(1, 10)]
    return results

if __name__ == "__main__":
    results = print_multiplication_table()
    print("\n".join(results)) if results != -1 else print(results) 
   
```

```python
import random

random_number = random.randint(1, 100)
guessed_number = 0

print("숫자를 맞춰보세요 (1 ~ 100)")

while True:
    guessed_number = int(input())

    if guessed_number > random_number:
        print("숫자가 너무 큽니다")
    elif guessed_number < random_number:
        print("숫자가 너무 작습니다")
    else:
        print(f"정답입니다. 입력한 숫자는 {guessed_number}입니다")
        break
```

```python
def multiplication_table() -> None:

    while True:
        print("구구단 몇 단을 계산할까요(1 ~ 9, 종료: 0)?")
        try:
            table = int(input())
            if table == 0:
                print("구구단 게임을 종료합니다.")
                break

            if table < 1 or table > 9:
                print("1부터 9까지 값을 입력하세요.")
                continue
        except ValueError as err:
            print(f"1부터 9까지 정수 값을 입력하세요.\n<<{err}>>")
            continue

        print(f"구구단 {table}단을 계산합니다.")    
        for i in range(1, 10):
            print(f"{table} X {i} = {table * i: >2}")

if __name__ == "__main__":
    multiplication_table()
```

```python
class MultiplicationTable():
    def __init__(self, table: int) -> None:
        try:
            self.table = int(table)

            if not (2 <= self.table <= 9):
                raise ValueError("2에서 9 사이의 정수값을 입력해 주세요.")
            
        except ValueError as err:
            print(f"입력값에 문제가 있습니다: {err}")
            self.table = None
            self.results = []
            return
        
        self.results = []
        self.calculate()

    def calculate(self) -> list:
        if self.table:
            self.results = [f"{self.table} X {i} = {self.table * i: 2d}" for i in range(1, 10)]
        return self.results
    
    def __str__(self) -> str:
       if not self.results:
           return "계산된 결과값이 없습니다. 입력값을 확인해 주세요."
       return "\n".join(self.results)
```

```python
g1 = MultiplicationTable(2)
print(g1)
```
