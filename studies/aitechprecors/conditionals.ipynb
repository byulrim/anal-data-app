```python
scores = [38, 37, 7, 16, 95, 71, 63, 48, 49, 66, 37]
grades = [""] * len(scores)

for i, score in enumerate(scores):
    if score >= 90:
        grade = "A"
    elif score >= 80:
        grade = "B"
    elif score >=70:
        grade = "C"
    elif score >= 60:
        grade = "D"
    else:
        grade ="F"
    
    grades[i] = grade

for score, grade in zip(scores, grades):
    print(f"{score: >2}: {grade}")
    
```

```python
print("당신이 태어난 년도를 입력하세요")
birth_year = int(input())
korean_age = 2026 - birth_year + 1

if 26 >= korean_age >= 20:
    print("대학생")
elif 20 > korean_age >= 17:
    print("고등학생")
elif 17 > korean_age >= 14:
    print("중학생")
elif 14 > korean_age >= 8:
    print("초등학생")
else:
    print("학생이 아닙니다.")
```

```python
print("구구단 몇 단을 계산할까요?")
table = int(input())

print(f"구구단 {table}단을 계산합니다.")

for i in range(1, 10):
    print(f"{table} X {i} = {table * i}")
```

```python
import random

random_number = random.randint(1, 100)
guessed_number = 0

print("숫자를 맞춰보세요 (1 ~ 100)")

while True:
    guessed_number = int(input())

    if guessed_number > random_number:
        print("숫자가 너무 큽니다")
    elif guessed_number < random_number:
        print("숫자가 너무 작습니다")
    else:
        print(f"정답입니다. 입력한 숫자는 {guessed_number}입니다")
        break
```

```python
def multiplication_table() -> None:

    while True:
        print("구구단 몇 단을 계산할까요(1 ~ 9, 종료: 0)?")
        try:
            table = int(input())
            if table == 0:
                print("구구단 게임을 종료합니다.")
                break

            if table < 1 or table > 9:
                print("1부터 9까지 값을 입력하세요.")
                continue
        except ValueError as err:
            print(f"1부터 9까지 정수 값을 입력하세요.\n<<{err}>>")
            continue

        print(f"구구단 {table}단을 계산합니다.")    
        for i in range(1, 10):
            print(f"{table} X {i} = {table * i: >2}")

if __name__ == "__main__":
    multiplication_table()
```
