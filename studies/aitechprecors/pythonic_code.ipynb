# split & join

```python
items = "zero one two three".split()
print(items, type(items))
```

```python
example = "python, java, javascript"
for language in example.split(","):
    print(language.strip())
```

```python
example = "teamlab.technology.io"
subdomain, domain, tld = example.split(".") # tld: top level domain
print(subdomain, domain, tld)
```

```python
colors = ["red", "blue", "green", "yellow"]
print("-".join(colors))
print(colors)
```

# list comprehension


```python
result = []
for i in range(10):
    result.append(i)
print(result)
```

```python
result = [i for i in range(10)]
print(result)
```

```python
word_1 = "Hello"
word_2 = "World"

result = []

for i in word_1:
    for j in word_2:
        result.append(i + j)
print(result)
```

```python
word_1 = "Hello"
word_2 = "World"

result = [i + j for i in word_1 for j in word_2]
print(result)
```
