```python
kor_scores = [49, 79, 20, 100, 80]
math_scores = [43, 59, 85, 30, 90]
eng_scores = [49, 79, 48, 60, 100]
midterm_scores = [kor_scores, math_scores, eng_scores]
print(midterm_scores)
```

```python
midterm_scores_copy = midterm_scores.copy()
print(midterm_scores_copy is midterm_scores)
print(midterm_scores_copy[0] is midterm_scores[0])
print(midterm_scores_copy == midterm_scores_copy)
```

```python
import copy

midterm_scores_copy = copy.deepcopy(midterm_scores)
print(midterm_scores_copy is midterm_scores)
print(midterm_scores_copy[0] is midterm_scores[0])
print(midterm_scores_copy == midterm_scores_copy)
```
