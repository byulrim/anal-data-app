```python
lst = [1, 2, 3, 4, 5]
print(lst)
lst.append(10) # push
print(lst)
lst.append(20)
print(lst)
a = lst.pop() # pop
print(lst)
print(a)
b = lst.pop()
print(lst)
print(b)
```

```python
lst = [1, 2, 3, 4, 5]
print(lst)
lst.append(10) # put
print(lst)
lst.append(20)
print(lst)
a = lst.pop(0) # get
print(lst)
print(a)
b = lst.pop(0)
print(lst)
print(b)
```
