```python
# stack
lst = [1, 2, 3, 4, 5]
print(lst)
lst.append(10) # push
print(lst)
lst.append(20)
print(lst)
a = lst.pop() # pop
print(lst)
print(a)
b = lst.pop()
print(lst)
print(b)
```

```python
word = input("Input a word: ")
word_lst = list(word)
print(word_lst)

for _ in range(len(word_lst)):
    print(word_lst.pop())
    print(word_lst)
```

```python
#queue
lst = [1, 2, 3, 4, 5]
print(lst)
lst.append(10) # put
print(lst)
lst.append(20)
print(lst)
a = lst.pop(0) # get
print(lst)
print(a)
b = lst.pop(0)
print(lst)
print(b)
```

```python
s = set([1, 2, 3, 1, 2, 3])
print(s)
s.add(1)
print(s)
s.remove(1)
print(s)
s.update([1, 4, 5, 6, 7])
print(s)
s.discard(8)
print(s)
s.clear()
print(s)
```

```python
s1 = {1, 2, 3, 4, 5}
s2 = {3, 4, 5, 6, 7}

print(s1.union(s2)) # 합집합
print(s1 | s2)

print(s1.intersection(s2)) # 교집합
print(s1 & s2)

print(s1.difference(s2)) # 차집합
print(s1 - s2)

print(s1.symmetric_difference(s2)) # 대칭 차집합
print(s1 ^ s2)
```
