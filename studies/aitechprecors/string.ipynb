```python
def factorial(n):
    if n == 1:
        return 1
    else:
        return n * factorial(n - 1)
    
if __name__ == "__main__":
    print(factorial(5))
```

```python
def factorial(number: int) -> int:
    """
    Calculate Factorial Number
    """
    if number == 1:
        return 1

    for i in range(number - 1, 1, -1):
        number *= i
    
    return number

print(factorial(5))
```
